## Categorical Columns (Text / Labels)
#### 1. Transaction ID
A unique code for each purchase.
Used to identify individual transactions.
Categorical because it represents an ID, not a number for calculation.

#### 2. Customer ID
A unique identifier for each customer.
Helps track customer behavior.
Categorical since it is an identifier, not a measurable value.

#### 3. Category
The product category (e.g., Clothing, Electronics).
Useful for grouping and analyzing broad product types.

#### 4. Item
The specific product purchased (e.g., “Shirt,” “USB Cable”).
Categorical because it represents product names.

#### 5. Payment Method
The method used to pay (e.g., Cash, Credit Card, Mobile Pay).
Helps understand customer payment preference patterns.

#### 6. Location
The store where the purchase took place.
Useful for comparing sales across branches.

#### 7. Transaction Date
The date when the transaction occurred.
Currently stored as text, but can be converted to datetime for time-based analysis.

#### 8. Discount Applied
Shows whether a discount was used (e.g., “Yes,” “No,” or value).
Categorical since it represents a status or label.

## Numerical Columns (Numbers used for calculations)
#### 9. Price Per Unit
The cost of a single unit of the purchased item.
Used to calculate total spending and price analysis.

#### 10. Quantity
Number of units bought in the transaction.
Indicates purchase volume and demand for items.

#### 11. Total Spent
The total amount spent in the transaction.
Often equal to Price Per Unit × Quantity.
Key metric for revenue analysis.

## detecting outliers based on all numeric columns

```python
numeric_cols=df.select_dtypes(include=['number']).columns
numeric_cols
```

```python
print("\n === Outliers Count(IQR method) ===")

for col in numeric_cols:
    Q1= df[col].quantile(0.25)
    Q3= df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col} : {len(outliers)} outliers")
```

##  Fix Inconsistencies in Categorical Data

```python
cat_cols = ['Transaction ID', 'Customer ID', 'Category', 'Item',
            'Payment Method', 'Location', 'Transaction Date']

for col in cat_cols:
    df[col] = df[col].astype(str).str.strip().str.title()

for col in cat_cols:
    df[col] = df[col].str.replace(r'\s+', ' ', regex=True)

df['Payment Method'] = df['Payment Method'].replace({
    'Creditcard': 'Credit Card',
    'Credit-Card': 'Credit Card',
    'Credit Card ': 'Credit Card',
    'Cash ': 'Cash',
    'Mobile Pay ': 'Mobile Pay',
    'Mobile Payment': 'Mobile Pay'
})

df['Location'] = df['Location'].str.replace('-', ' ')
df['Location'] = df['Location'].str.title().str.strip()

df['Category'] = df['Category'].replace({
    'Electonics': 'Electronics',
    'Clothes': 'Clothing',
    'Kid Wear': 'Kids Wear',
    'Kid’S Wear': 'Kids Wear'
})

df.head()
```

## Convert Data Types

## Convert Transaction Date to datetime

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
```

##  Convert numerical columns to numeric types

```python
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
```

##  Convert categorical columns to category type

```python
cat_cols = ['Transaction ID', 'Customer ID', 'Category', 'Item',
            'Payment Method', 'Location']

for col in cat_cols:
    df[col] = df[col].astype('category')
```
